```python
# import dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# import csv file
csv_file = 'Resources/TA_hotels_scraper.csv'
TA_hotels_df = pd.read_csv(csv_file, encoding='latin1')
TA_hotels_df.head()
```

```python
# transform
TA_hotels_df = TA_hotels_df.rename(columns={"continent_name": "Continent", "city_name": "City", "country_name": "Country", "Price": "Price (USD)"})
TA_hotels_df["Price (USD)"] = round(TA_hotels_df["Price (USD)"], 2)
TA_hotels_df.head()
```

```python
# connect to local database
rds_connection_string = "postgres:postgres@localhost:5432/TA_hotels_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
TA_hotels_df.to_sql(name='TA_hotels', con=engine, if_exists='append', index=False)
```
